```python
import os
import pandas as pd
# We'll also import seaborn, a Python graphing library
import warnings # current version of seaborn generates a bunch of warnings that we'll ignore
warnings.filterwarnings("ignore")
import seaborn as sns
import matplotlib.pyplot as plt
sns.set(style="white", color_codes=True)
home = os.path.expanduser('~')
%matplotlib inline
```

```python
train_df =  pd.read_csv(home + '/datalab/Rossman/train.csv')
test_df  = pd.read_csv(home + '/datalab/Rossman/test.csv')
store_df =  pd.read_csv(home + '/datalab/Rossman/store.csv')                       
```

```python
## sales depending on day of week 
sns.FacetGrid(train_df, hue="DayOfWeek", size=10) \
   .map(plt.scatter, "Sales", "Customers") \
   .add_legend()
```

```python


# We can look at the sales in Seaborn through a boxplot
sns.boxplot(x="DayOfWeek", y="Sales", data=train_df)

```

```python
# We can look at an individual feature in Seaborn through a boxplot
sns.boxplot(x="DayOfWeek", y="Customers", data=train_df)
```

```python
# We can look at an individual feature in Seaborn through a boxplot
sns.boxplot(x="Promo", y="Customers", data=train_df)
```

```python
# We can look at an individual feature in Seaborn through a boxplot
sns.boxplot(x="Promo", y="Sales", data=train_df)
```

```python


# A final seaborn plot useful for looking at univariate relations is the kdeplot,
# which creates and visualizes a kernel density estimate of the underlying feature
sns.FacetGrid(train_df, hue="Promo", size=6) \
   .map(sns.kdeplot, "Sales") \
   .add_legend()

```
